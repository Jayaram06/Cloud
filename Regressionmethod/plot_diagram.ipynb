```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
data1 = pd.read_csv("diabetes.csv")

# Extract BMI and Age columns
bmi = data1['BMI']
age = data1['Age']

# Plot BMI vs Age
plt.figure(figsize=(8, 6))
plt.scatter(age, bmi, color='blue', alpha=0.5)
plt.title('BMI vs Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.grid(True)
plt.show()
```

```python
data1 = pd.read_csv("iris.csv")

# Extract sepal length and width columns
sepal_length = data1['SepalLengthCm']
sepal_width = data1['SepalWidthCm']

# Plot BMI vs Age
plt.figure(figsize=(8, 6))
plt.scatter(sepal_length, sepal_width, color='blue', alpha=0.5)
plt.title('sepal_length vs sepal_width')
plt.xlabel('sepal_width')
plt.ylabel('sepal_length')
plt.grid(True)
plt.show()
```
